```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('income per capita')
plt.scatter(df.year,df.income_per_capita,color='red',marker='+')
```

```python
reg = LinearRegression()
```

```python
reg.fit(df[['year']], df.income_per_capita)
```

```python
reg.predict([[2015]])
```

```python
reg.predict([[2030]])
```
